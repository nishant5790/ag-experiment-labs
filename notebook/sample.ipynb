```python
# # build web search agent

server_url = "https://research-mcp-9mm5.onrender.com/mcp/mcp"
async def run_agent(message: str) -> None:
    # Initialize and connect to the Streamable HTTP MCP server
    mcp_tools = MCPTools(transport="streamable-http", url=server_url)
    await mcp_tools.connect()

    try:
        agent = Agent(
            name="search_agent",
            model=Gemini(
                id= "gemini-2.5-flash"
            ),
            tools=[mcp_tools],
            instructions='use sources to search on topic and share the relevant source and links for the given topic '


        )

        await agent.aprint_response(input=message, stream=True, markdown=True)
    finally:
        await mcp_tools.close()
```

```python
# if __name__ == "__main__":
#     asyncio.run(run_agent("ai engineering"))

await run_agent("ai engineering")
```

```python
# calling agent
stream: Iterator[RunOutputEvent] = search_agent.run("Latest AI news?", stream=True)
for chunk in stream:
    if chunk.event == RunEvent.run_content:
        print(chunk)
```
